```pyspark
from pyspark.sql.functions import udf
```

```pyspark
def price_range(unitPrice):
    avg = 4.611113626088354;
    
    if unitPrice > (avg + avg*0.50):
        return 'Higher than avarage'
    elif unitPrice < (avg - avg*0.50):
        return 'Lower than Avarage'
    else:
        return 'Btw the avarage'
```

```pyspark
brand_udf = udf(price_range, StringType())

df.withColumn('price_range', brand_udf(df['UnitPrice'])).select('StockCode','Description','Quantity','UnitPrice',"price_range").show(10,False)
```

```pyspark
quantity_udf = udf(lambda quantity: "Greater than avg" if quantity <= 9.55224954743324 else "less than avg", StringType())

df.withColumn("Quantity_insight",quantity_udf(df.Quantity)).select('StockCode','Description','Quantity','UnitPrice',"Quantity_insight").show(10,False)
```

```pyspark
from pyspark.sql.functions import pandas_udf, PandasUDFType
```

```pyspark
def missing_to_the_goal(Quantity):
    Quantity_goal = 50;
    
    missing_to_goal = Quantity_goal - Quantity
    
    if missing_to_goal <= 0:
        return 0;

    return missing_to_goal;
```

```pyspark
missing_to_the_goal_udf = udf(missing_to_the_goal, IntegerType())
df.withColumn("missing_to_quantity_goal", missing_to_the_goal_udf(df['Quantity'])).select('StockCode','Quantity','UnitPrice',"missing_to_quantity_goal").show(10,False)
```

```pyspark
def earned(Quantity,UnitPrice):
    x=Quantity*UnitPrice
    return x
```

```pyspark
earned_udf = udf(earned, DoubleType())

df.withColumn("earned", earned_udf(df['Quantity'],df['UnitPrice'])).select('CustomerID','StockCode','Quantity','UnitPrice',"earned").show(10,False)
```
